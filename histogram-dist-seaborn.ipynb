# Histogram and Distribution Plot

### A Histogram is a univariate plot which helps us understand the distribution of a continuous numerical variable.

### A Distribution plot can show univariate and bivariate distributions.

```python
import pandas as pd
import seaborn as sns
```

### Import diamonds dataset

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
sns.displot(diamonds['price'], kind='hist')
```

```python
sns.displot(data=diamonds,x = 'price', kind='kde')
```

```python
sns.displot(data=diamonds, x='price', kind='ecdf')
```

```python
sns.displot(data=diamonds, x='price', rug=True, kind='kde')
```

### Price and carat

```python
sns.displot(data=diamonds, x='price', y='carat') # unclear distribution
```

```python
sns.histplot(data = diamonds, x='price')
```

### Color parameter

```python
sns.histplot(data=diamonds, x='price', color='green')
```

### Change axis and number of bins

### give the column name to y axis, set number to a bins parameter

```python
sns.histplot(data=diamonds, y='price', color='red', bins=20)
```

## Add a kernal density estimate with kde parameter set to True

```python
sns.histplot(data=diamonds, x='price', kde=True)
```

## If no parameter is assigned, neither x nor y is assigned, the dataset is treated as wide-form, and a histogram is drawn for each numeric column:

```python
# sns.histplot(data=diamonds)
```
