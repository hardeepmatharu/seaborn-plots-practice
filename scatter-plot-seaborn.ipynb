# Scatter Plot

### When you want to know the potential relationship between two variables, change in one affects the other variable. 

### Scatter plot is an analytical tool that shows a statistical relationship between two continuous variables.

### Positive correlation: when value y tends to increase as the x variable increases, it is positive correlation.

### Negative correlation: when value of y tends to decrease as x variable increases, it is negative correlation.

```python
import pandas as pd
import seaborn as sns
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
sns.scatterplot(data=iris, x='SepalLengthCm', y='PetalLengthCm')
```

### It shows positive correation

### hue parameter with categorical column

```python
sns.scatterplot(data=iris, y='PetalLengthCm', x='SepalLengthCm', hue='Species')
```

### hue parameter with numeric column

```python
sns.scatterplot(data=iris, x='SepalLengthCm',y='PetalLengthCm', hue='PetalLengthCm')
```

### It shows the intensity of the PetallenghtCm as SepalLengthCm grows
