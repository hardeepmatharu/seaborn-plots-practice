# Line Plot in Seaborn

### Line plots are used for dataset where it has multiple y values for each x value, for data that evolves over time. for ex. data over months in several years. This is also called seasonality analysis.

### It is also called dot plot, as straight line connects dots on the graph.

### Line plots are created to emphasize the changes in value, rather than the magnitude of the values themselves.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal')
plt.show()
```

### Unlike line plot in matplotlib, line plot in seaborn shows the axis labels

## Hue parameter, differentiates the line based on column categories given in the 'hue'

```python
sns.lineplot(data = fmri, x='timepoint', y='signal', hue='region')
plt.show()
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event')
plt.show()
```

## Line aesthetics, based on column in hue parameter

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='event')
plt.show()
```

## Show markers on the lines

```python
sns.lineplot(data=fmri, x='timepoint',y='signal',hue='event', markers=True, style='event')
plt.show()
```

## Confidence interval

### Area around the line,shaded area, is called confidence interval.

### Confidence interval is a range of values around that estimate that are believed to contain the true values of that estimate with a certain probability.

### ci parameter - which is set to False to remove confidence interval

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', errorbar=('ci',False))
plt.show()
```

## Limitations: 
 ### 1. Clustered: when there is large dataset, too lines make it clustered, difficult to read.
 ### 2. Fraction/Decimal Data: if there are fractions in the dataset then it is inconvenient to display.
