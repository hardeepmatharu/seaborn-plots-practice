# Pair Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Distrbution of all numeric columns
### Diagonal graphs will have itself distribution

```python
mydata = pd.read_csv('Iris.csv')
```

```python
mydata.head()
```

```python
sns.pairplot(mydata)
```

### hue parameter for more clarity

```python
sns.pairplot(mydata, hue='Species')
```

### Set diagonal graphs to histograms

```python
sns.pairplot(mydata,hue='Species', diag_kind = 'hist')
```

### For all graphs of kind 'kde'

```python
sns.pairplot(mydata, hue='Species', kind='kde')
```

### All graphs of kind 'hist'

```python
sns.pairplot(mydata, kind='hist', hue='Species')
```

### Get rid of all the repetitive graphs, all above the diagonal

```python
sns.pairplot(mydata, hue='Species', corner=True)
```

### Change markers, 'o' for circle, 's' for square, 'D' for diamond

```python
sns.pairplot(mydata, hue='Species', markers=['o','s','D'])
```
