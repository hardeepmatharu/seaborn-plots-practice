# Joint Plot

## Joint plot helps to quickly visualize the bivariate and univariate profiles on the same plot.

## It comprises of three plots - one displays the relationship between two variables while the other two show the individual distribution of each variable in the margins.

## By default, joint plot creates a scatter plot to show the relationship between two variables and individual histograms are created in x and y axis margins for each of the variables.

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mydata = pd.read_csv('Iris.csv')
```

```python
mydata.head()
```

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm')
```

### Show scale for both histograms 

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm', marginal_ticks=True)
```

### color parameter, which changes the color of the graphs, hue parameter to differentiate the data

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm', color='brown', hue='Species')
```

### Show regression line, by setting 'kind' parameter to 'reg'  

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm', color='brown', kind='reg')
```

### scatter plot is not useful for large dataset due to overlapping of points so 'hex' kind is used to visualize the data

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm', color='blue', kind='hex')
```

### Try kind = kde

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm', hue='Species', kind='kde')
```

### Add color bar to see the density of the data

```python
sns.jointplot(data=mydata, x='SepalLengthCm', y='PetalLengthCm', color='green', kind='hex')
plt.colorbar()
```
