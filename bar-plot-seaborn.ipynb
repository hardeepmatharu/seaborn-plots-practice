# Bar Plot 

## Bar plot is used for categorical data and numeric data.It shows the average of the numeric data against categorical data.

```python
import seaborn as sns
import pandas as pd
```

```python
mydata = pd.read_csv('pokemon.csv')
```

```python
mydata.head()
```

```python
sns.set(style='whitegrid')
```

```python
sns.barplot(x='is_legendary',y='speed', data=mydata)
```

```python
mydata[mydata['is_legendary'] == 0]
```

```python
mydata.loc[:,['speed','is_legendary']] # only 2 columns using loc 
```

## hue parameter, it differentiates based on data in the hue parameter column

```python
sns.barplot(data=mydata, x='is_legendary', y='speed', hue='generation')
```

## palette parameter, used to assign color to bars

```python
sns.barplot(data=mydata, x='is_legendary',hue='is_legendary', y='speed', palette='Blues_d')
```

### Try 'rocket' palette

```python
sns.barplot(data=mydata, x='is_legendary', y='speed', hue='is_legendary', palette='rocket')
```

### Try 'vlag' palette

```python
sns.barplot(data=mydata, x='is_legendary', y='speed', hue='is_legendary', palette='vlag')
```

### Custom color for all the bars

```python
sns.barplot(data=mydata, x='is_legendary', y='speed', color='orange')
```
