# Box Plot

### A box plot or a box-and-whisker plot shows the distribution of numerical data and skewness through displacing the data quartiles.

### It shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.

### It is also called five number summary as it contains the five things to describe the data, min, Q1,median,Q3 and max. 

### Whiskers length: if right length is shorter then it is left skewed or negative skewed.

### Whiskers lenght: if left length is shorter then it is right skewed or positive skewed.

```python
import pandas as pd
import seaborn as sns
```

```python
mydata = pd.read_csv('churn.csv')
```

```python
mydata.head()
```

```python
sns.boxplot(data=mydata, x='Churn', y='Tenure')
```

### There are equal chances that the customer will churn, leave the organization.

```python
sns.boxplot(data=mydata, x='Gender', y='Tenure')
```

### Equal tenure for male and female 

```python
sns.boxplot(data=mydata, x='Gender', y='Total Spend')
```

### male and female have equal median but female has more distribution from min to max.

### hue parameter is used to differentiate more clearly

```python
sns.boxplot(data=mydata, x='Gender', y='Total Spend', hue='Subscription Type')
```

### Change the order of columns, use order parameter

```python
sns.boxplot(data=mydata, x='Gender', y='Total Spend', order=['Male','Female'], hue='Subscription Type')
```

### Change line width, set linewidth to some number to increase or decrease the thickness of the lines and can change the width of the boxes using widht parameter

```python
sns.boxplot(data=mydata, x='Gender', y='Tenure', linewidth=3, width=0.5)
```
